```python
from sklearn import preprocessing
def convert(data):
    number = preprocessing.LabelEncoder()
    for col_name in data.columns:
        data[col_name] = number.fit_transform(data[col_name])
        data=data.fillna(-999)
    return data
```

# H. Overview of each Section using simple Bar graphs

The significance of features as calculated by a Random Forest classifier are shown in this barplot. To help with feature selection and to understand the aspects that significantly impact diabetes categorization, it shows the relative value of each feature in predicting the target variable (class). This helps with data exploration by drawing attention to important factors that may be used to develop models.

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

X = df.drop('class', axis=1)
y = df['class']

rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X, y)

feature_importances = pd.Series(rf_classifier.feature_importances_, index=X.columns)
feature_importances.sort_values(ascending=False, inplace=True)

plt.figure(figsize=(10, 8))
sns.barplot(x=feature_importances.index, y=feature_importances.values, palette='magma')
plt.title('Feature Importances')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# I. Histograms
In this step we use numeric feature histograms to display dataset distribution and attributes. Histograms show data structure, relevance, quality, pattern identification, and engineering. The dataset's central tendency, distribution, and shape are shown. The unique attribute distributions in histograms help forecast target variables. Learn numeric feature distribution to increase predictive modeling accuracy and interpretability.

```python
numeric_features = df.select_dtypes(include=['float64', 'int64']).columns

fig, axes = plt.subplots(nrows=len(numeric_features), ncols=1, figsize=(12, 10))

for i, feature in enumerate(numeric_features):
    sns.histplot(df[feature], ax=axes[i], kde=True, color='skyblue')
    axes[i].set_title(f'Histogram of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')

plt.tight_layout()

plt.suptitle('Histograms of Numeric Features', x=0.5, y=1.02, ha='center', fontsize='large')

plt.show()

```

# J. Correlation Matrix
By plotting the correlations between the numerical characteristics, we can see how they relate to one another and how they rely on one another. In a positive correlation, the two variables are moving in the same direction; in a negative correlation, the opposite is true. To improve the accuracy of diabetes analysis predictions, it is helpful to understand these correlations. This knowledge helps with feature selection and model creation.

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

# K. Scatter Plot
Age and PG concentration are shown in this scatter figure, which helps to understand how they are related in the dataset. One way to gain a better grasp of how demographic factors impact medical indicators in diabetes analysis is to visually examine the plot. # K. Pairplot
The pairplot shows scatter plots of the characteristics that were chosen, with markers separating the class labels. It provides insights into correlations and class separability by graphically representing pairwise interactions and distributions within the dataset. This helps in comprehending how features interact with one another and adds to thorough data exploration, which in turn informs judgments about feature selection and predictive modeling.

```python
selected_features = ['Diastolic BP' ,'Pregnancies', 'PG Concentration', 'class']

sns.pairplot(df[selected_features], hue='class', markers=['o', 's'], palette='Set1')
plt.suptitle('Pairplot of Selected Features', y=1.02, ha='center', fontsize='large', fontweight='bold')
plt.show()
```
